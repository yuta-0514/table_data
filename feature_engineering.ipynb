```python
import pandas as pd

data = {"date":
        ["2017-11-01 12:24",
        "2017-11-18 23:00",
        "2017-12-05 5:05",
        "2017-12-22 8:54",
        "2018-01-08 14:20",
        "2018-01-19 20:01"
        ],
        "sales": [200, 220, 250, 270, 300, 310]
}

df = pd.DataFrame(data)
```

```python
# 年月日と時間をそれぞれを抽出
def add_date(df, col):
  df[col] = pd.to_datetime(df[col])
  df["year"] = df["date"].dt.year
  df["month"] = df["date"].dt.month
  df["day"] = df["date"].dt.day
  return df

# n個前のデータを追加
def add_lag_shift(df, col, n):
  df[f"{col}_lag_{n}"] = df[col].shift(n)
  return df
df = add_lag_shift(df, "sales", 1)

# n周期の過去window個の移動平均などを追加
def add_lag_moving_ave(df, col, window, n):
  df[f"{col}_moving_ave_{window}_{n}"] = df[col].shift(n).rolling(window=window).mean()
  df[f"{col}_moving_max_{window}_{n}"] = df[col].shift(n).rolling(window=window).max()
  df[f"{col}_moving_min_{window}_{n}"] = df[col].shift(n).rolling(window=window).min()
  df[f"{col}_moving_median_{window}_{n}"] = df[col].shift(n).rolling(window=window).median()
  return df
```
