```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

data = pd.read_csv('https://raw.githubusercontent.com/mwaskom/seaborn-data/master/iris.csv')
data['species'] = data['species'].replace('setosa', 1)
data['species'] = data['species'].replace('versicolor', 2)
data['species'] = data['species'].replace('virginica', 3)
```

```python
# 基礎統計量確認
data.describe()
```

```python
# ヒストグラム
data.hist()

# 歪度
data['species'].skew()

# 尖度
data['species'].kurt()
```

```python
# 箱ひげ図
data.plot.box()
```

```python
# 散布図
data.plot.scatter(x='sepal_length', y='species')
```

```python
# ヒートマップ
data.corr().style.background_gradient(axis=None)
```
