```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

data = pd.read_csv('https://raw.githubusercontent.com/mwaskom/seaborn-data/master/iris.csv')
data['species'] = data['species'].replace('setosa', 1)
data['species'] = data['species'].replace('versicolor', 2)
data['species'] = data['species'].replace('virginica', 3)
```

```python
# 基礎統計量確認
data.describe()
```

```python
# ヒストグラム
data.hist()
```

```python
# 箱ひげ図
sns.boxplot(x=data['sepal_length'])
```

```python
# 散布図
sns.pairplot(data)
```

```python
# カテゴリごとの散布図
sns.pairplot(data, hue='species')
```

```python
# ヒートマップ
sns.heatmap(data)
```
