# Stratified k-fold
クラスの割合を均等にする

```python
import pandas as pd

data = pd.read_csv('https://raw.githubusercontent.com/mwaskom/seaborn-data/master/iris.csv')
```

```python
import numpy as np
from sklearn.model_selection import StratifiedKFold

train_x = data[['sepal_length', 'sepal_width', 'petal_length', 'petal_width']]
train_y = data['species']

kf = StratifiedKFold(n_splits=5, shuffle=True)
for tr_idx, va_idx in kf.split(train_x, train_y):
  tr_x, va_x = train_x.iloc[tr_idx], train_x.iloc[va_idx]
  tr_y, va_y = train_y.iloc[tr_idx], train_y.iloc[va_idx]
```

### 時系列データの場合

```python
import pandas as pd
import plotly.express as px

df = px.data.stocks()
df['year'] = df['date'].apply(lambda x :x.split("-")[0])
df['year'] = df['year'].astype(int)
df['month'] = df['date'].apply(lambda x :x.split("-")[1])
df['month'] = df['month'].astype(int)

train_x = df.drop(columns='GOOG')
train_y = df['GOOG']

# ４，５，６月がVali、それ以前の3ヶ月がTrain
va_period_list = [4, 5, 6]
train_width = 3
for vaperiod in va_period_list:
  is_tr = (train_x['month'] < vaperiod) & (train_x['month'] >= vaperiod-train_width)
  is_va = train_x['month'] == vaperiod
  tr_x, va_x = train_x[is_tr], train_x[is_va]
  tr_y, va_y = train_y[is_tr], train_y[is_va]

```
